## web scraping of amazon best selling book


we scrape the data of book name, writer name, star of book, book link, star of book and book price.

```python
import requests
```

```python
header = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/104.0.5112.81 Safari/537.36 Edg/104.0.1293.47'}
```

```python
topics_url = 'https://www.amazon.in/gp/bestsellers/books/'
```

```python
response = requests.get(topics_url,headers = header)
```

```python
response.status_code
```

```python
len(response.text)
```

```python
page_contents = response.text
```

```python
from bs4 import BeautifulSoup
```

```python
doc = BeautifulSoup(page_contents, 'html.parser')
```

# Book Name

```python
div_tag = doc.find_all('div',class_="_cDEzb_p13n-sc-css-line-clamp-1_1Fn1y")
```

```python
book_name = []
writer_name = []
name_list = []
for image in div_tag:
    name_list.append(image.text)
    
    
```

```python
book_name.append(name_list[::2])
writer_name.append(name_list[1::2])
```

```python
for title in book_name:
    print(title)
    
```

# Writer Name

```python
for writer in writer_name:
    print(writer)
```

```python
len(writer)
```

# Book Link

```python
link = doc.find_all('a',class_="a-link-normal",tabindex ='-1')
```

```python
link_list = []
for link_tag in link:
    link_list.append(topics_url+link_tag['href'])
```

```python
for book_link in link_list:
    print(book_link)
```

# making data frame

```python
topics_dict = {
    'book_title': title,
    'writer': writer,
    'url': book_link
}
```

```python
import pandas as pd
```

```python
topic_df =  pd.DataFrame(topics_dict)
topic_df
```

# star of book

```python
div_tag2  = doc.find_all('span',class_="a-icon-alt")
```

```python
len(div_tag2)
```

```python
star_list = []
for star in div_tag2:
    star_list.append(star.text)
```

```python
len(star_list)
```

```python
topic_df['star'] = star_list
```

```python
topic_df
```

# price of book

```python
price = doc.find_all('span',class_="a-size-base a-color-price")
```

```python
price_list = []
for price_tag in price:
    price_list.append(price_tag.text)
```

```python
len(price_list)
```

```python
topic_df['price'] = price_list
```

```python
topic_df
```

```python
topic_df['price'] = topic_df['price'].str.replace('₹', '')
topic_df['price'] = topic_df['price'].astype(float)
```
